```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('datasets/new_dataset.csv', header = None)
data
```

```python
size = len(data.index)
min_sup = int(size/30)
print('size = ', size)
print('min_sup = ', min_sup)
```

```python
support=data.apply(pd.value_counts)
support = support.drop([-2, -1], axis=0)
support['support'] = support.sum(axis=1)
support['set'] = support.index
support = support.loc[support['support'] > min_sup]
support = pd.DataFrame().assign(set=support['set'],support=support['support'])
support =support.reset_index(drop=True)
print(support)
```

**so there are 9 unique values greater than minSup**
